```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

# 한글 폰트 설정
fm._get_fontconfig_fonts()
font_location = '/Users/user/Library/Fonts/D2Coding-Ver1.3.2-20180524-all.ttc'
font_name = fm.FontProperties(fname=font_location).get_name()
matplotlib.rc('font', family=font_name)

# 파일을 다시 불러와 막대그래프를 그립니다. 평균선과 중앙값선을 추가합니다.
data = pd.read_csv('주요관광지점_입장객_2020-2022.csv')

# 막대 그래프 그리기
plt.figure(figsize=(10, 8))
plt.bar(data['관광지'], data['총계'])

# 평균 및 중앙값 계산
mean_val = data['총계'].mean()
median_val = data['총계'].median()

# 평균선 및 중앙값선 그리기
plt.axhline(y=mean_val, color='r', linestyle='-', label=f'평균: {mean_val:.2f}')
plt.axhline(y=median_val, color='g', linestyle='-', label=f'중앙값: {median_val:.2f}')

# 레이블, 타이틀 및 범례 설정
plt.xlabel('관광지')
plt.ylabel('총계')
plt.title('2020~22년 관광지별 입장객 수 (막대그래프)')
plt.xticks(rotation=90)
plt.legend()

# 그래프 출력
plt.show()

# 평균 및 중앙값 출력
mean_val, median_val

```

```python
# 선 그래프로 변경하여 그립니다. 평균선과 중앙값선을 그대로 유지합니다.

plt.figure(figsize=(10, 8))
plt.plot(data['관광지'], data['총계'], marker='o')  # 선 그래프 생성, 각 지점에 마커 표시

# 평균선 및 중앙값선 그리기
plt.axhline(y=mean_val, color='r', linestyle='-', label=f'평균: {mean_val:.2f}')
plt.axhline(y=median_val, color='g', linestyle='-', label=f'중앙값: {median_val:.2f}')

# 레이블, 타이틀 및 범례 설정
plt.xlabel('관광지')
plt.ylabel('총계')
plt.title('관광지별 입장객 수 (선 그래프)')
plt.xticks(rotation=90)
plt.legend()

# 그래프 출력
plt.show()
```

```python
# 막대 그래프와 선 그래프를 함께 그려보겠습니다.

plt.figure(figsize=(10, 8))

# 막대 그래프
plt.bar(data['관광지'], data['총계'], alpha=0.6)  # 약간 투명하게 설정하여 선 그래프가 보이도록 함

# 선 그래프
plt.plot(data['관광지'], data['총계'], color='b', marker='o')  # 선 그래프 생성, 각 지점에 마커 표시

# 평균선 및 중앙값선 그리기
plt.axhline(y=mean_val, color='r', linestyle='-', label=f'평균: {mean_val:.2f}')
plt.axhline(y=median_val, color='g', linestyle='-', label=f'중앙값: {median_val:.2f}')

# 레이블, 타이틀 및 범례 설정
plt.xlabel('관광지')
plt.ylabel('총계')
plt.title('2020~22년 관광지별 입장객 수')
plt.xticks(rotation=90)
plt.legend()

# 그래프 출력
plt.show()
```
